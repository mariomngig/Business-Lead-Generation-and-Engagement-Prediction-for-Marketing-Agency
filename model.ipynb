```python
import pandas as pd
```

```python
df = pd.read_csv("ig_solar - Copy.csv",index_col=False)
```

```python
df
```

```python
df1 = pd.read_csv("ig_home_remodeling - Copy.csv")
df1
```

```python
df2 = pd.read_csv("ig_plumping - Copy.csv",index_col=False)
df2
```

```python
final_df = pd.concat([df,df1,df2],axis=0,ignore_index=True)
final_df
```

```python
final_df.fillna(0,inplace=True)
final_df
```

```python
data = ['followers','last_post_likes','last_post_comments','dif_days',"same_business_idea"]
```

```python
X = final_df[data]
y = final_df['reach_out']
```

```python
X
```

```python
from sklearn.preprocessing import LabelBinarizer
lb = LabelBinarizer()
X['same_business_idea'] = lb.fit_transform(X["same_business_idea"])
X 
```

```python
y = lb.fit_transform(y)
y = y.ravel()
y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,stratify=X["same_business_idea"],test_size=0.2,random_state=42)

```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

# Check the class distribution before and after SMOTE
print("Original class distribution:", Counter(y_train))
print("Resampled class distribution:", Counter(y_train_res))
```

```python
from sklearn.ensemble import RandomForestClassifier 

rdtc = RandomForestClassifier(max_depth=3,random_state=42,oob_score=True,criterion='entropy')
rdtc.fit(X_train_res,y_train_res)
```

```python
rdtc.score(X_train_res,y_train_res)
```

```python
rdtc.oob_score_
```

```python
from sklearn.metrics import precision_score , recall_score

y_pred = rdtc.predict(X_train_res)

print(precision_score(y_train_res,y_pred))
print(recall_score(y_train_res,y_pred))
```

```python
y_pred = rdtc.predict(X_test)

print(precision_score(y_test,y_pred))
print(recall_score(y_test,y_pred))
```

```python
rdtc.score(X_test,y_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state=42)
ada= AdaBoostClassifier(dtc,random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
# Define the base estimator
dtc = DecisionTreeClassifier(random_state=42)

# Define the parameter grid
param_grid = {
    'base_estimator__max_leaf_nodes': [2,4,6],
    'base_estimator__max_depth': [2, 3, 4],
    'base_estimator__splitter': ['best', 'random'],
    'base_estimator__min_samples_leaf': [1, 5, 10],
    'base_estimator__min_samples_split': [2, 10, 20],
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 1]
}
```

```python
grid_search = GridSearchCV(estimator=ada, param_grid=param_grid, cv=3, n_jobs=-1, scoring='precision')
```

```python
grid_search.fit(X_train_res,y_train_res)
```

```python
best_params = grid_search.best_params_
best_estimator = grid_search.best_estimator_
best_score = grid_search.best_score_

print("Best Parameters:", best_params)
print("Best Estimator:", best_estimator)
print("Best Score:", best_score)
```

```python
from sklearn.model_selection import cross_val_score
# Assume X and y are your features and target data
scores = cross_val_score(grid_search.best_estimator_, X_train_res, y_train_res, cv=4, scoring='precision')
```

```python
print(f"Cross-validated scores: {scores}")
print(f"Mean cross-validated score: {scores.mean()}")
print(f"Standard deviation of cross-validated scores: {scores.std()}")
```

```python
y_pred = grid_search.best_estimator_.predict(X_train_res)

print(precision_score(y_train_res,y_pred))
print(recall_score(y_train_res,y_pred))
```
